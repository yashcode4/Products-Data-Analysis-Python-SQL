```python
# Extract file from zip file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() # extract file to dir
zip_ref.close() # close file
```

```python
# Read data from the file adn handle null values
import pandas as pd
df = pd.read_csv('orders.csv')
df.head(20)
```

```python
df['Ship Mode'].unique()
```

```python
# Removing 'Not Available', 'unknown' values from the feature
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
df.head(1)
```

```python
# Rename column names. Make them lower case and replace space with underscore
# df.rename(columns={'Order Id': 'order_id', 'City': 'city'})
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head(1)
```

```python
# Derive new columns discount, sale price and profit
df['discount'] = df['list_price']*df['discount_percent']*.01
df['sale_price'] = df['list_price']-df['discount']
df['profit'] = df['sale_price']-df['cost_price']
```

```python
df.head()
```

```python
# Convert order date from object data types to datetime
df.dtypes
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
```

```python
df.dtypes
```

```python
# Drop cost_price, list_price and discount_percent columns
df.drop(columns=['list_price', 'cost_price', 'discount_percent'], inplace=True)
```

```python
# Load the data into sql server using replace option
# !pip install sqlalchemy psycopg2
import sqlalchemy as sal
engine = sal.create_engine('postgresql+psycopg2://postgres:1234@localhost:5432/postgres')
conn = engine.connect()
```

```python
# Load the data into postgresql server using append option
df.to_sql('df_orders', con=conn, index=False, if_exists='append')
```
